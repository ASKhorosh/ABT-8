# Гипотеза 1
Если мы перестанем показывать информацию о надбавке за повышенный спрос для тех, кто едет не из центра, в случае запроса цены на поездку, то конверсия в заказ на шаге показа цены увеличится на 14%, потому что среди данной группы конверсия 82% среди тех, кто не видит отметку о повышенном спросе и 68% среди тех, кто видит пометку.  
**Генеральная совокупность:** Пользователи, совершающие поездки не из центра

### Шаги АВ-теста: 
1. Гипотеза сформулирована (записана выше)
2. В качестве исследуемой метрики берем *конверсию в завершение поездки из тех, кто видел цену без пометки*. В качестве метрики, которая не должна ухудшиться, берем *конверсию в отмену заказа из тех, кто заказал* (т.е. манипуляции с ценой не должны повышать уровень отмен уже после заказа, когда пользователь пожалел, что заказал по этой цене, с пометкой она или нет) 
3. Считаем размер выборок http://joxi.ru/Y2L1b9KHMW7YWA  В нашем случае размер выборки позволяет провести такой тест
4. Готовим выборку (об этом ниже) 
5. С помощью установленного в компании инструмента для АВ-тестирования (покупного, либо собственного), настраиваем тест: делаем варианты отображения метки повышенного спроса, загружаем тем или иным способом выборку, проверяем, что  нужная нам статистика для подведения итогов корректно собирается и сохраняется, настраиваем сроки проведения теста и иные параметры, которые требует наш инструмент.
6. Запускаем тест, ждем сбора данных

```python
#Готовим выборку
no_city_center = ab.loc[(ab['city_center_order'] == 0)]
no_city_center
```

```python
#берем только тех, кто не видел пометку
no_city_center = no_city_center.loc[(no_city_center['surge'] == 'no surge')]
no_city_center
```

```python
#считаем показатели конверсию в завершение поездки по юзерам среди тех, кто увидел цену без пометки, т.к. именно на эту метрику мы хотим влиять
no_city_center_group.price_seen = no_city_center.groupby('user_id', as_index=False)['price_seen'].sum()['price_seen']
no_city_center_group.ride_completed = no_city_center.groupby('user_id', as_index=False)['ride_completed'].sum()['ride_completed']
no_city_center_group['r_c/p_s'] = np.where(no_city_center_group['price_seen'] < 1, no_city_center_group['price_seen'], no_city_center_group['ride_completed']/no_city_center_group['price_seen'])
no_city_center_group
```

```python
#удаляем значения, где конверсия не рассчитана
no_city_center_group = no_city_center_group.dropna()
no_city_center_group
```

```python
import random
pd.options.mode.chained_assignment = None
no_city_center_group['randNumGroup'] = np.random.randint(1,3, size=len(no_city_center_group))
no_city_center_group
```

```python
no_city_center_group1 = no_city_center_group.loc[(no_city_center_group['randNumGroup'] == 1)]
no_city_center_group1
```

```python
no_city_center_group2 = no_city_center_group.loc[(no_city_center_group['randNumGroup'] == 2)]
no_city_center_group2
```

```python
print (len(no_city_center_group1) - len (no_city_center_group2))
```

```python
import scipy.stats as stats
#попробуем сравнить группы по показателю завершения поездки
t, p_value = stats.ttest_ind(no_city_center_group1['r_c/p_s'], no_city_center_group2['r_c/p_s']) #Критерий Стьюдента (t-test)
t, p_value
```

```python
stats.ttest_ind(no_city_center_group1['r_c/p_s'], no_city_center_group2['r_c/p_s'], equal_var = False)
```

```python
alpha = 0.05
if p_value < alpha:
    print('p-value меньше уровня значимости -- значит гипотезу о том, что выборки различаются, следует отвергнуть')
else: 
    print('p-value больше уровня значимости -- значит гипотезу о том, что выборки различаются, следует принять') 
```
